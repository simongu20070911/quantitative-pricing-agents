```python
import databento as db

client = db.Historical("YOUR_API_KEY")

# Check size BEFORE downloading
bytes_count = client.timeseries.get_billable_size(
    dataset='GLBX.MDP3',
    symbols=['ES.c.0'],        # Your Smart Symbol
    stype_in='continuous',     # Smart Logic
    schema='ohlcv-1m',         # 1-minute bars
    start='2010-01-01',
    end='2020-01-01'
)

# Convert to Gigabytes (GB)
gb_count = bytes_count / (1024 * 1024 * 1024)

# Databento Historical CME data is usually around $25 - $30 per GB
estimated_cost = gb_count * 30 

print(f"Size: {gb_count:.4f} GB")
print(f"Estimated Cost: ${estimated_cost:.2f}")
```
